```python
#import libraries
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib notebook
```

## Inflation Data Visualization

```python
readcsv=("Data/Inflation.csv")
S_W_Inflation_df=pd.read_csv(readcsv)
S_W_Inflation_df.head()
```

```python
#Create a line chart of inflation (above) and home sales per year from 1963-2018
home_sales=[]
south_sales=[]
west_sales=[]
for sale in S_W_Inflation_df["United States Home Sales"]:
    int_US_sales=int(sale)
    home_sales.append(int_US_sales)
for sale in S_W_Inflation_df["South Home Sales"]:
    int_south_sales=int(sale)
    south_sales.append(int_south_sales)
for sale in S_W_Inflation_df["West Home Sales"]:
    int_west_sales=int(sale)
    west_sales.append(int_west_sales)
fig, ax1=plt.subplots(figsize=(15,7))
x=np.arange(1963,2018)
n=S_W_Inflation_df["Inflation Rate"]
ax1.plot(x,n, linewidth=2.5, color="black")
ax2=ax1.twinx()
ax2.plot(x,S_W_Inflation_df["United States Home Sales"],color="r",label="National Home Sales")
ax2.plot(x,S_W_Inflation_df["South Home Sales"], color="g", label="South Home Sales")
ax2.plot(x,S_W_Inflation_df["West Home Sales"], color="olive", label="West Home Sales")
ax1.grid()
ax1.legend(loc=2)
ax2.legend(loc=1)
ax1.set_title("Inflation v. Home Sales ", fontweight="bold")
ax1.set_xlabel("Year", fontweight="bold")
ax1.set_ylabel("Inflation (%)", fontweight="bold")
ax2.set_ylabel("Home Sales", fontweight="bold")
plt.savefig("Graph Images/Inflation_Home_Sales.png")
```

```python
# Plot Inflation rate Vs National Home Sales
home_sales=[]
for sale in S_W_Inflation_df["United States Home Sales"]:
    int_US_sales=int(sale)
    home_sales.append(int_US_sales)
fig, ax3=plt.subplots(figsize=(15,7))
x=np.arange(1963,2018)
n=S_W_Inflation_df["Inflation Rate"]
ax3.plot(x,n, linewidth=2.5, color="black")
ax4=ax3.twinx()
ax4.plot(x,S_W_Inflation_df["United States Home Sales"],color="r",label="National Home Sales")
ax3.grid()
ax3.legend(loc=2)
ax4.legend(loc=1)
ax3.set_title("Inflation v. Home Sales (Natl)", fontweight="bold")
ax3.set_xlabel("Year", fontweight="bold")
ax3.set_ylabel("Inflation (%)", fontweight="bold")
ax4.set_ylabel("Home Sales", fontweight="bold")
plt.savefig("Graph Images/Inflation_National.png")
```

```python
#Plot Inflation Rate Vs South Home Sales
south_sales=[]
for sale in S_W_Inflation_df["South Home Sales"]:
    int_south_sales=int(sale)
    south_sales.append(int_south_sales)
fig, ax5=plt.subplots(figsize=(15,7))
x=np.arange(1963,2018)
n=S_W_Inflation_df["Inflation Rate"]
ax5.plot(x,n, linewidth=2.5, color="black")
ax6=ax5.twinx()
ax6.plot(x,S_W_Inflation_df["South Home Sales"],color="g",label="South Home Sales")
ax5.grid()
ax5.legend(loc=2)
ax6.legend(loc=1)
ax5.set_title("Inflation v. Home Sales (South)", fontweight="bold")
ax5.set_xlabel("Year", fontweight="bold")
ax5.set_ylabel("Inflation (%)", fontweight="bold")
ax6.set_ylabel("Home Sales", fontweight="bold")
plt.savefig("Graph Images/Inflation_South.png")
```

```python
#Plot Inflation Rate Vs West Home Sales
west_sales=[]
for sale in S_W_Inflation_df["West Home Sales"]:
    int_west_sales=int(sale)
    west_sales.append(int_west_sales)
fig, ax7=plt.subplots(figsize=(15,7))
x=np.arange(1963,2018)
n=S_W_Inflation_df["Inflation Rate"]
ax7.plot(x,n, linewidth=2.5, color="black")
ax8=ax7.twinx()
ax8.plot(x,S_W_Inflation_df["West Home Sales"],color="olive",label="West Home Sales")
ax7.grid()
ax7.legend(loc=2)
ax8.legend(loc=1)
ax7.set_title("Inflation v. Home Sales (West)", fontweight="bold")
ax7.set_xlabel("Year", fontweight="bold")
ax7.set_ylabel("Inflation (%)", fontweight="bold")
ax8.set_ylabel("Home Sales", fontweight="bold")
plt.savefig("Graph Images/Inflation_West.png")
```

```python
inflation_post_TwoK= S_W_Inflation_df[42:55]
inflation_pre_TwoK = S_W_Inflation_df[0:42]
n_groups=3
fig, ax7 = plt.subplots()

post_2004_data=[inflation_post_TwoK['South Home Sales'].corr(inflation_post_TwoK['Inflation Rate']),
inflation_post_TwoK['West Home Sales'].corr(inflation_post_TwoK['Inflation Rate']),
inflation_post_TwoK['United States Home Sales'].corr(inflation_post_TwoK['Inflation Rate'])]

pre_2004_data=[inflation_pre_TwoK['South Home Sales'].corr(inflation_pre_TwoK['Inflation Rate']),
inflation_pre_TwoK['West Home Sales'].corr(inflation_pre_TwoK['Inflation Rate']),
inflation_pre_TwoK['United States Home Sales'].corr(inflation_pre_TwoK['Inflation Rate'])]

index=np.arange(n_groups)
bar_width = 0.35

rects1 = ax7.bar(index, pre_2004_data, bar_width, color='r',alpha=.75)

rects2 = ax7.bar(index + bar_width, post_2004_data, bar_width,color='g', alpha=.75)
ax7.set_xlabel("Home Sale Category",fontweight="bold")
ax7.set_ylabel('Correlation Coefficient',fontweight="bold")
ax7.set_title('Inflation x Home Sales Correlation, Pre and Post 2004',fontweight="bold")
ax7.set_xticklabels(('South Home Sales', 'West Home Sales', 'United States Home Sales'))
ax7.legend((rects1[0], rects2[0]),('Pre 2004', 'Post 2004'), loc="lower left")
ax7.grid(axis="y")
ax7.set_xticks(index + bar_width / 2)
ax7.set_ylim(top=1, bottom =-1)
rects = ax7.patches
y_values = pre_2004_data + post_2004_data

for rect, y_ in zip(rects, y_values):
    height = rect.get_height()
    if height >0:
        ax7.text(rect.get_x() + rect.get_width() / 2., height + 0.05, str((round(height,2))) ,ha='center', va='bottom',color='k')
    else:
        ax7.text(rect.get_x() + rect.get_width() / 2., height - 0.1, str((round(height,2))) ,ha='center', va='bottom',color='k')
    


fig.tight_layout()
plt.savefig('Graph Images/Inflation_Home_Sales_Correlation.png')
```

## Unemployment Data Visualization

```python
unemployment_df = pd.read_csv("Data/unemployment.csv")
unemployment_df.head()
```

```python
natl_home_sales=[]
for sale in unemployment_df["United States Home Sales"]:
    int_natl_sale=int(sale)
    natl_home_sales.append(int_natl_sale)
fig, ax1=plt.subplots(figsize=(15,10))
x=np.arange(1980,2016)
n=unemployment_df['National Unemployment Rate']
ax1.plot(x,n, color='indigo', label="National Unemployment Rate", linewidth=2.5)
ax2=ax1.twinx()
ax2.plot(x,natl_home_sales,color="r", label='Natl Home Sales')
ax1.grid()
ax1.set_title('Unemployment v. Natl Home Sales (1980-2015)',fontweight="bold")
ax1.set_xlabel('Year',fontweight="bold")
ax1.set_ylabel('Unemployment (%)',fontweight="bold")
ax2.set_ylabel('Home Sales (000s)',fontweight="bold")
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")
plt.savefig('Graph Images/Unemployment_Home_Sales.png')
```

```python
south_sales=[]
for sale in unemployment_df["South Home Sales"]:
    int_south_sale=int(sale)
    south_sales.append(int_south_sale)
fig, ax3=plt.subplots(figsize=(15,7))
x=np.arange(1980,2016)
s=unemployment_df['South Unemployment Rate']
ax3.plot(x,s,color="indigo", label="South Unemployment Rate", linewidth=2.5)
ax4=ax3.twinx()
ax4.plot(x,south_sales,color="g", label='South Home Sales')
ax3.grid()
ax3.set_title('Unemployment v. South Home Sales (1980-2015)',fontweight="bold")
ax3.set_xlabel('Year',fontweight="bold")
ax3.set_ylabel('Unemployment (%)',fontweight="bold")
ax4.set_ylabel('Home Sales (000s)',fontweight="bold")
ax3.legend(loc="upper left")
ax4.legend(loc="upper right")
plt.savefig('Graph Images/Unemployment_South_Sales.png')
```

```python
west_sales=[]
for sale in unemployment_df["West Home Sales"]:
    int_west_sale=int(sale)
    west_sales.append(int_west_sale)
fig, ax5=plt.subplots(figsize=(15,7))
x=np.arange(1980,2016)
w=unemployment_df['West Unemployment Rate']
ax5.plot(x,w,color="indigo", label="West Unemployment Rate", linewidth=2.5)
ax6=ax5.twinx()
ax6.plot(x,west_sales,color="olive", label='West Home Sales')
ax5.grid()
ax5.set_title('Unemployment v. West Home Sales (1980-2015)',fontweight="bold")
ax5.set_xlabel('Year',fontweight="bold")
ax5.set_ylabel('Unemployment (%)',fontweight="bold")
ax6.set_ylabel('Home Sales (000s)',fontweight="bold")
ax5.legend(loc="upper left")
ax6.legend(loc="upper right")
plt.savefig('Graph Images/Unemployment_West_Sales.png')
```
